```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()   ##--> object yuazıyorsa categoric
            ## --> diğer veriler numeric
```

```python
sns.countplot(x="origin", data=df); ## hangisinden kac tane var onu görürüz
```

```python
sns.countplot(x="cylinders", data=df); ## en cok araba 4 silindirli
```

```python
sns.barplot(x="origin",y="horsepower", data=df) ##barplot da istatistik bir veri hesaplanır
                                                ##üstündeki çizgiden anlarız.
```

```python
sns.barplot(x="cylinders", y="horsepower", data=df);
```

```python
labels = df.origin.unique()
sizes = df.origin.value_counts()
explode = (0, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots(figsize=(6,6))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
sns.histplot(x="horsepower", data=df); ## bu grafik yayılım verir 50-225 arasında
                                    ## arabalar 50-75 arasında beygir var
```
